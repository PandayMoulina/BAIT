In this step, features are extracted from the URLs dataset.

```python
#importing required packages for the following section
from urllib.parse import urlparse, urlencode
import ipaddress
import re
```

### 3.1 URL starting with http or https

HTTPS is a strong indicator of legitimacy as data shared between browsers are encrypted in SSL/TSL, while HTTP often signals outdated or potentially unsafe sites.

However, note: HTTPS does not guarantee trustworthiness by itself—phishing or scam sites can still use HTTPS. But lack of HTTPS is a red flag.

If url starts with https it is marked as 1 else 0.


```python
def urlstart(url):
  sch = urlparse(url).scheme
  if sch=='https':
    return 1
  else:
    return 0
```

###3.2. Domain of the URL

Extracts the domain present in the URL.
A domain is a part of a website's address that identifies it on the internet. It serves as a unique identifier, allowing users to access a website by typing in a memorable name, such as 'example.com'.

Each domain is registered and must be unique. Phishers often create domains that closely resemble legitimate ones (eg. typosquatting), making it essential to scrutinize the domain for authenticity and for indicating potential phishing attempts.

```python
def getDomain(url):
  domain = urlparse(url).netloc
  if re.match(r"^www.",domain):
      domain = domain.replace("www.","")
  return domain
```



Here we decided to remove the prefix 'www.' from the Domain name in order to normalize and standardize the domain name. This helps us in the following ways:
1. Ensures consistency in comparison or storage of domain names. For example, 'www.example.com' and 'example.com' refer to the same website, but they are technically different strings.
2. Users may not always remember to type "www." when entering a web address. By normalizing the domain, the code can handle user input more effectively.
3. This will help focus on the core domain which makes it easier to spot discrepencies and aid in identifying phishiing attempts where attackers might
create a similar-looking domain.

### 3.3 IP Address in the URL


Checks for the presence of IP address in the URL. URLs may have IP address instead of domain name. Presence of IP addresses in URL instead of a domain name is alarming in the following ways:
1. IP addresses do not convey any recognizable brand or identity, making it harder for users to assess their legitimacy. Thus, increasing the risk of users unknowingly visiting a malicious site.
2. Attackers prefer to put IP addresses in the phishing URL to bypass security systems, including firewalls and intrusion detection systems, which are configured to monitor and block known blacklisted or malicious domains.
3. Domains can be blacklisted once identified as malicious, but switching to a new IP address is relatively easy for attackers. They can continuously change their IPs while maintaining the same malicious intent, thus prolonging their operations.
4. Registering a domain name can take time and may require verification processes, while IP addresses can often be used immediately without such hurdles. This allows phishers to quickly set up malicious sites before they are detected.








So if instead of any domain name an IP address is present in any URL, we can be certain that the URL is malicious and someone is trying to steal personal information with this URL.


```python
def havingIP(url):
  try:
    ipaddress.ip_address(url)
    ip=1
  except:
    ip=0
  return ip
```

If the domain part of URL has IP address, the value assigned to this feature is 1 (phishing) or else 0 (legitimate).

###3.4. "@" Symbol in URL

Checks for the presence of "@" symbol in the URL. Using "@" symbol in the URL leads the browser to ignore everything preceding the "@" symbol and the real address often follows after the "@" symbol. This allows attackers to obscure the true destination of the URL and thus confuse and mislead users. For example malicious links can be crafted in a similar pattern like“http://gmail.com@malicious_URL.com“. This is same as "www.malicious_URL.com" . Thus links having "@" symbol potentially leads to incorrect routing.





```python
def haveAtSign(url):
  if "@" in url:
    at=1 #indicates presence of '@' in URL
  else:
    at=0 #indicated the absence of '@' in URL
  return at
```

###3.5 Length of URL

This computes the length of the URL. Phishers can use long URL's to hide the doubtful part in the address bar in the following ways:

1.   Phishers often place a legitimate-looking subdomain or keywords at the start of the URL, making it look authentic. The actual malicious domain (where the attack is hosted) is hidden deeper in the URL and out of immediate view. For example:
https://www.login-secure-verification.paypal.com.fakedomain.com/secure_login

In this case, users might see "login-secure-verification.paypal.com" at first glance and miss the "fakedomain.com" further down the URL.


2. Phishers take the advantage of the above method as most modern browsers have limited display space in their URL bars and often truncate long URLs, showing only the beginning. Thus, the real domain or malicious content may be hidden beyond what is immediately visible.

3. Also, using long URLs having random charecters, or parameters are used to distract users from the malicious domain. For example:
https://legit-site.com/redirect?x=some-long-string&redirect=https://malicious.com





In this project, if the length of the URL is greater than or equal to 54 charecters then the URL is classified as phishing otherwise legitimate.

If the length of URL>=54, the value assigned to this feature is 1(phishing) or else 0(legitimate).

```python

def getLength(url):
  if len(url) < 54:
    length = 0
  else:
    length = 1
  return length
```

###3.6 Depth of URL

This code computes the depth of the URL. This feature calculates the number of sub pages in the given url based on the number of '/'.

The value of feature is a number based on '/' in the URL. For example:
1. https://example.com → Depth = 0
2. https://example.com/user/profile/settings → Depth = 3

```python
def getDepth(url):
  s = urlparse(url).path.split('/')
  depth = 0
  for j in range(len(s)):
    if len(s[j]) != 0:
      depth = depth+1
  return depth
```

###3.7 Redirection "//" in URL

Checks the presence of "//" in the URL. The existence of “//” within the URL path means that the user will be redirected to another website. The location of the “//” in URL is computed. We find that if the URL starts with “HTTP”, that means the “//” should appear in the sixth position. However, if the URL employs “HTTPS” then the “//” should appear in seventh position.

Phishers often use redirection (//) in URLs to deceive users by manipulating how URLs appear and behave. They place the legitimate-looking domain before the // to trick users. For example:
http://login.example.com//phishing-site.com/login

Here, the deceptive Part is 'login.example.com' appears trustworthy, bur the actual destination of this url is 'phishing-site.com'. The browser redirects to 'phishing-site.com'.

If the "//" is anywhere in the URL apart from after the protocol, the value assigned to this feature is 1 (phishing) or else 0 (legitimate).

```python
def redirection(url):
  pos = url.rfind('//')
  if pos > 6:
    if pos > 7:
      return 1
    else:
      return 0
  else:
    return 0
```

###3.8. "http/https" in Domain name

Checks for the presence of "http/https" in the domain part of the URL. The phishers may add the “HTTPS” token to the domain part of a URL in order to trick users and create a false sense of security. For example: In "http://https-login.example.com", the domain part is "https-login.example.com".

If the URL has "http/https" in the domain part, the value assigned to this feature is 1 (phishing) or else 0 (legitimate).

```python
def httpDomain(url):
  domain = urlparse(url).netloc
  if 'https' in domain:
    return 1
  else:
    return 0
```

###3.9. Using URL Shortening Services “TinyURL”

URL shortening is a method on the “World Wide Web” in which a URL may be made considerably smaller in length and still lead to the required webpage. This is accomplished by means of an “HTTP Redirect” on a domain name that is short, which links to the webpage that has a long URL.

This is important for phishing detection as phishers use URL shorteners to hide malicious links as the shortened URL conceals the true destinations and also encourage more users to click as short URLs look cleaner and more trustworthy.

If the URL is using Shortening Services, the value assigned to this feature is 1 (phishing) or else 0 (legitimate).

```python
#listing shortening services
shortening_services = r"bit\.ly|goo\.gl|shorte\.st|go2l\.ink|x\.co|ow\.ly|t\.co|tinyurl|tr\.im|is\.gd|cli\.gs|" \
                      r"yfrog\.com|migre\.me|ff\.im|tiny\.cc|url4\.eu|twit\.ac|su\.pr|twurl\.nl|snipurl\.com|" \
                      r"short\.to|BudURL\.com|ping\.fm|post\.ly|Just\.as|bkite\.com|snipr\.com|fic\.kr|loopt\.us|" \
                      r"doiop\.com|short\.ie|kl\.am|wp\.me|rubyurl\.com|om\.ly|to\.ly|bit\.do|t\.co|lnkd\.in|db\.tt|" \
                      r"qr\.ae|adf\.ly|goo\.gl|bitly\.com|cur\.lv|tinyurl\.com|ow\.ly|bit\.ly|ity\.im|q\.gs|is\.gd|" \
                      r"po\.st|bc\.vc|twitthis\.com|u\.to|j\.mp|buzurl\.com|cutt\.us|u\.bb|yourls\.org|x\.co|" \
                      r"prettylinkpro\.com|scrnch\.me|filoops\.info|vzturl\.com|qr\.net|1url\.com|tweez\.me|v\.gd|" \
                      r"tr\.im|link\.zip\.net"
```

```python
def tinyURL(url):
    match=re.search(shortening_services,url)
    if match:
        return 1
    else:
        return 0
```

###3.10. Prefix or Suffix "-" in Domain

Checking the presence of '-' in the domain part of URL. The dash symbol is rarely used in legitimate URLs. Phishers tend to add prefixes or suffixes separated by (-) to the domain name in order to imitate trusted brands while bypassing domain restrictions so that users feel that they are dealing with a legitimate webpage.

If the URL has '-' symbol in the domain part of the URL, the value assigned to this feature is 1 (phishing) or else 0 (legitimate).

```python
# 9.Checking for Prefix or Suffix Separated by (-) in the Domain (Prefix/Suffix)
def prefixSuffix(url):
    if '-' in urlparse(url).netloc:
        return 1            # phishing
    else:
        return 0            # legitimate
```

### 3.11 No of dots (.) in URL domain

Every dot (.) separates parts of a domain or path. While some dots are normal, too many dots—especially in the domain—can be a red flag.

Legitimate domains usually have 2 or 3 dots (subdomain + domain + TLD).

Suspicious URLs may have 4 or more dots to mask the real domain and look trustworthy.

More dots often mean deep subdomain nesting, which attackers use to confuse users.

```python
def countDots(url):
    domain = urlparse(url).netloc
    c = domain.count('.')
    if c<3:
        return 0
    else:
      return 1
```

#4. Computing URL Features

Here we create a list and a function that calls the other functions and stores all the features of the URL in the list. We will extract the features of each URL and append to this list.

```python
#Function to extract features
def featureExtraction(url,label):

  features = []

  features.append(getDomain(url))
  features.append(urlstart(url))
  features.append(havingIP(url))
  features.append(haveAtSign(url))
  features.append(getLength(url))
  features.append(getDepth(url))
  features.append(redirection(url))
  features.append(httpDomain(url))
  features.append(tinyURL(url))
  features.append(prefixSuffix(url))
  features.append(countDots(url))
  features.append(label)
  return features
```

###4.1. Legitimate URLs:

Now, feature extraction is done on legitimate URLs.

```python
legiurl.shape
```

```python
#Extracting the feautres & storing them in a list
legi_features = []
label = 0

for i in range(0, 5000):
  url = legiurl['URL'][i]
  fea=featureExtraction(url,label)
  legi_features.append(fea)
```

```python
feature_names = ['Domain', 'http/https','Have_IP', 'Have_At', 'URL_Length', 'URL_Depth','Redirection',
                      'https_Domain', 'TinyURL', 'Prefix/Suffix','dots', 'Label']

legitimate = pd.DataFrame(legi_features, columns= feature_names)
legitimate.head()
```
